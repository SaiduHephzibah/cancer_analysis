## adaboost(adaptive boosting)

develope classifier model subsequently and reduce randome 
and biased ness by cutting dow continue until all data is unbiased

```python
# perform adaptive boosting

from sklearn import model_selection
from sklearn.ensemble import AdaBoostClassifier
seed=7 # intial point starting point
num_trees=30
kfold=model_selection.KFold(n_splits=10,random_state=seed) # increasing or decreasing splits causes reduced acuracy (overfitting/underfitting)
model=AdaBoostClassifier(n_estimators=num_trees,random_state=seed)
results=model_selection.cross_val_score(model,x_train,y_train,cv=kfold) # comparing accuracy
```

```python
results
```

```python
print(results.mean())
```
